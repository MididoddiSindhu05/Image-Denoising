```python

import matplotlib.pyplot as plt
import numpy as np


# File path
file_path = '/content/band2_image2.rad'

# Set correct dimensions
height, width = 512, 1024  # Adjusted based on file size

# Read file as binary and convert to uint8
with open(file_path, 'rb') as f:
    img_data = np.frombuffer(f.read(), dtype=np.uint8)

# Reshape to 2D image
image = img_data.reshape((height, width))

# Display
plt.imshow(image, cmap='gray')
plt.title('RAD Image Visualization (512x1024)')
plt.axis('off')
plt.show()
```

```python

import cv2  # OpenCV
from google.colab.patches import cv2_imshow

# Gaussian Blur
denoised_image = cv2.GaussianBlur(image, (5, 5), 0)

# Show denoised image
plt.imshow(denoised_image, cmap='gray')
plt.title('Denoised Image (Gaussian Blur)')
plt.axis('off')
plt.show()
```
